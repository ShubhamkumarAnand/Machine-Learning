#### Matplotlib Workflow

```python
#Import Matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare The Data
x = [1,5,6,8,10,12]
y = [20,30,50,60,90,110]

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(10,10)) #Width and Height

# 3. Plot Data
ax.plot(x,y)

# 4. Customise Plot
ax.set(title = "Simple Plot",
      xlabel = "x-axix",
      ylabel = "y-axis")

# 5. Save and Show 
fig.savefig("img/simple-plot.png")
```

#### OO Method for Plotting the Various Graph : Advanced Graph Plotting

```python
over_50 =heart_disease[heart_disease["age"] > 50] 
over_50.head()
```

```python
# Pyplot Method of Graph

over_50.plot(x="age",y="chol",kind="scatter",c="target")
```

```python
## OO Method with PyPlot Method

fig , ax = plt.subplots(figsize=(10,7))

over_50.plot(x="age",y="chol",kind="scatter",c="target",ax=ax);
#ax.set_xlim([60,75]);
```

```python
## OO Method From Scratch
fig,ax = plt.subplots(figsize = (10,6))

## Plot the Data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

## Customize the plot
ax.set(title="Heart Disease vs Cholestrol Leval",
      xlabel ="Age",
      ylabel = "Cholestrol");

## Add a legend Element
ax.legend(*scatter.legend_elements(),title="Target");

## Adding a Horizontal Line
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
over_50.head()
```

```python
# Subplot of age , thalach , trestbps
fig, (ax0,ax1) = plt.subplots(nrows=2, ncols=1,figsize = (10,10),sharex=True)

# For ax0
# Plot the Data 
scatter = ax0.scatter(x=over_50["age"],y=over_50["thalach"], c=over_50["target"]);

# Customize the Graph
ax0.set(title="Max Heart Rate vs Age", ylabel = "Max Heart Rate");

#Add a Label to the Target
ax0.legend(*scatter.legend_elements(),title="Target")

#Add a Mean line to the y-axis
ax0.axhline(over_50["thalach"].mean(),linestyle = "--");

#for ax1
#plot the Data
scatter = ax1.scatter(x=over_50["age"],y=over_50["trestbps"],c=over_50["target"])

#Customize the Graph
ax1.set(title =" BPS vs Age", xlabel ="Age",ylabel = "BP/S");

# Adding a Target
ax1.legend(*scatter.legend_elements(),title ="Target")

# Adding a MeanLine
ax1.axhline(over_50["trestbps"].mean(),linestyle = "--");

#Adding a Title
fig.suptitle("Heart Disease Ananlysis", fontsize = 18, fontweight = "bold")
```

```python
plt.style.use("seaborn-whitegrid")

## OO Method From Scratch
fig,ax = plt.subplots(figsize = (10,6))

## Plot the Data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                    cmap="winter") # This changes Color

## Customize the plot
ax.set(title="Heart Disease vs Cholestrol Leval",
      xlabel ="Age",
      ylabel = "Cholestrol");

## Add a legend Element
ax.legend(*scatter.legend_elements(),title="Target");

## Adding a Horizontal Line
ax.axhline(over_50["chol"].mean(),linestyle="--",color='r');
```

```python
# Custimizing the x and y axix of the plot
# Subplot of age , thalach , trestbps
fig, (ax0,ax1) = plt.subplots(nrows=2, ncols=1,figsize = (10,10),sharex=True)

# For ax0
# Plot the Data 
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter");

# Customize the Graph
ax0.set(title="Max Heart Rate vs Age", ylabel = "Max Heart Rate");

# Change the x & y axix of the plot
ax0.set_xlim([50,80])
ax0.set_ylim([50,200])

#Add a Label to the Target
ax0.legend(*scatter.legend_elements(),title="Target")

#Add a Mean line to the y-axis
ax0.axhline(over_50["thalach"].mean(),linestyle = "--",color='r');

#for ax1
#plot the Data
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["trestbps"],
                      c=over_50["target"],
                      cmap="winter")

# Change the x & y axix of the plot
ax1.set_ylim([80,220])

#Customize the Graph
ax1.set(title =" BPS vs Age", xlabel ="Age",ylabel = "BP/S");

# Adding a Target
ax1.legend(*scatter.legend_elements(),title ="Target")

# Adding a MeanLine
ax1.axhline(over_50["trestbps"].mean(),linestyle = "--",color='r');

#Adding a Title
fig.suptitle("Heart Disease Ananlysis", fontsize = 18, fontweight = "bold");
```
